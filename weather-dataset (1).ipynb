```python
#feature formation
features=['datetime_utc', ' _conds', ' _dewptm', ' _fog', ' _hail',
       ' _heatindexm', ' _hum', ' _precipm', ' _pressurem', ' _rain', ' _snow',
        ' _thunder', ' _tornado', ' _vism', ' _wdird', ' _wdire',
       ' _wgustm', ' _windchillm', ' _wspdm']
```

```python
#defining features matrix X and prediction value y
X=data[features]
y=data[' _tempm']
```

```python
X.head()
```

```python
#since the feature 'datetime_utc' does not affect the Discharge Pressure prediction and therefore it can be drop
X.drop(['datetime_utc'],inplace=True,axis=1)
#since the feature ' _precipm' is unavailable in dataset and therefore it can be drop
X.drop([' _precipm'],inplace=True,axis=1)
```

```python
X.head()
```

```python
X.shape
```

```python
#checking if any missing data in our dataset
data.isnull().sum()
```

```python
X[' _conds'].fillna(method='ffill',inplace=True)
X[' _wdire'].fillna(method='ffill',inplace=True)
```

```python
X.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
le=LabelEncoder()
X[' _conds']=le.fit_transform(X[' _conds'].astype('str'))
X[' _wdire']=le.fit_transform(X[' _wdire'].astype('str'))
#ohe=OneHotEncoder(categorical_features=[0,13])
#X=ohe.fit_transform(X).toarray()
```

```python
X.head()
```

```python
X.isnull().sum()
```

```python


from sklearn.impute import SimpleImputer

imputer = SimpleImputer(missing_values= np.nan, strategy='mean')
X = imputer.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

```python
#Fitting model
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor(n_estimators=10,random_state=100)
model.fit(X_train,y_train)
y_pred_train=model.predict(X_train)
y_pred_test=model.predict(X_test)
```

```python
#checking accuracy of model using rmse value on training dataset
from sklearn.metrics import mean_squared_error
from math import sqrt

rms = sqrt(mean_squared_error(y_train, y_pred_train))
print(rms)
```

```python
#checking accuracy of model using rmse value on test dataset
rms = sqrt(mean_squared_error(y_test, y_pred_test))
print(rms)
```

**rmse value can be minimizes by increasing n_estimators value**
